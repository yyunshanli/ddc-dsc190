````python
def prepare_context_window(features, context_radius=7):
    # Add padding for context
    padded = np.pad(features,
                    ((context_radius, context_radius), (0, 0), (0, 0)),
                    mode='constant')
    
    # Create sliding windows
    windows = []
    for i in range(context_radius, len(features) + context_radius):
        # Take 7 frames before and after current frame
        window = padded[i-context_radius:i+context_radius+1]
        windows.append(window)
        
    return np.array(windows)
````

````python
import tensorflow as tf
import numpy as np

def show_cnn_flow(windows):
    """
    Demonstrate how data flows through CNN layers
    windows shape: (batch_size, 15, 80, 3)
    """
    print(f"Input shape: {windows.shape}")
    
    # First CNN block
    x = tf.keras.layers.Conv2D(
        filters=10,              # Number of output filters
        kernel_size=(7, 3),      # Size of the convolution window
        padding='same'           # Padding to keep spatial dimensions
    )(windows)
    print(f"\nAfter first Conv2D: {x.shape}")
    # Shape is now (batch_size, 15, 80, 10)
    
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    # Shape unchanged by BatchNorm and ReLU
    
    x = tf.keras.layers.MaxPooling2D(
        pool_size=(1, 3)         # Pool only in frequency dimension
    )(x)
    print(f"After first MaxPool2D: {x.shape}")
    # Shape is now (batch_size, 15, 27, 10)
    
    # Second CNN block
    x = tf.keras.layers.Conv2D(
        filters=20,              # More filters in second layer
        kernel_size=(3, 3),      
        padding='same'
    )(x)
    print(f"\nAfter second Conv2D: {x.shape}")
    # Shape is now (batch_size, 15, 27, 20)
    
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    
    x = tf.keras.layers.MaxPooling2D(
        pool_size=(1, 3)
    )(x)
    print(f"After second MaxPool2D: {x.shape}")
    # Shape is now (batch_size, 15, 9, 20)
    
    return x

# Example usage:
batch_size = 32
windows = np.random.random((batch_size, 15, 80, 3))  # Example batch of windows

# Create full model to demonstrate
class StepDetectionCNN(tf.keras.Model):
    def __init__(self):
        super(StepDetectionCNN, self).__init__()
        
        # First CNN block
        self.conv1 = tf.keras.layers.Conv2D(10, (7, 3), padding='same')
        self.bn1 = tf.keras.layers.BatchNormalization()
        self.pool1 = tf.keras.layers.MaxPooling2D((1, 3))
        
        # Second CNN block
        self.conv2 = tf.keras.layers.Conv2D(20, (3, 3), padding='same')
        self.bn2 = tf.keras.layers.BatchNormalization()
        self.pool2 = tf.keras.layers.MaxPooling2D((1, 3))
        
        # Final layers for step detection
        self.flatten = tf.keras.layers.Flatten()
        self.dense1 = tf.keras.layers.Dense(256, activation='relu')
        self.dense2 = tf.keras.layers.Dense(1, activation='sigmoid')
    
    def call(self, inputs):
        # First block
        x = self.conv1(inputs)
        x = self.bn1(x)
        x = tf.nn.relu(x)
        x = self.pool1(x)
        
        # Second block
        x = self.conv2(x)
        x = self.bn2(x)
        x = tf.nn.relu(x)
        x = self.pool2(x)
        
        # Final layers
        x = self.flatten(x)
        x = self.dense1(x)
        x = self.dense2(x)
        
        return x

# Example training loop
def train_step(model, optimizer, x_batch, y_batch):
    """Single training step"""
    with tf.GradientTape() as tape:
        # Forward pass
        predictions = model(x_batch, training=True)
        # Calculate loss
        loss = tf.keras.losses.binary_crossentropy(y_batch, predictions)
        
    # Calculate gradients
    gradients = tape.gradient(loss, model.trainable_variables)
    # Apply gradients
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    
    return loss

def prepare_labels(chart, context_radius=7):
    """
    Convert chart annotations into binary labels
    
    Args:
        chart: OnsetChart object containing annotations and dt (time step)
        context_radius: Same context radius used for feature windows
    
    Returns:
        Binary labels array aligned with feature windows
    """
    # Get total number of frames from song features
    num_frames = len(chart.song_features)
    
    # Create empty labels array
    labels = np.zeros(num_frames)
    
    # Fill in 1s where steps occur
    for annotation in chart.annotations:
        # annotation[1] contains time in seconds
        # chart.dt is the time step (usually 0.01 seconds)
        frame_idx = int(annotation[1] / chart.dt)
        
        # Make sure we don't go out of bounds
        if frame_idx < num_frames:
            labels[frame_idx] = 1
            
    return labels
````

````python
def open_dataset_fps(*args):
    print(args)
    un_tuple_args = [i[0] if type(i) == tuple else i for i in args]
    datasets = []
    print(un_tuple_args)
    for data_fp in un_tuple_args:
        if not data_fp:
            datasets.append([])
            continue
        print(data_fp)
        with open(data_fp, 'r') as f:
            song_fps = f.read().split()

        dataset = []

        for song_fp in song_fps:
            with open(song_fp, 'rb') as f:
                data = CustomUnpickler(f).load()
                # dataset.append(pickle.load(f))
                dataset.append(data)

        datasets.append(dataset)
    return datasets[0] if len(datasets) == 1 else datasets
````

````python
def load_and_preprocess_data(train_fp, valid_fp, test_fp, z_score, experiment_dir):
    train_data, valid_data, test_data = open_dataset_fps(train_fp, valid_fp, test_fp)
    if z_score:
        z_score_fp = os.path.join(experiment_dir, 'valid_mean_std.pkl')
        if valid_fp and not os.path.exists(z_score_fp):
            mean_per_band, std_per_band = calc_mean_std_per_band(valid_data)
            with open(z_score_fp, 'wb') as f:
                pickle.dump((mean_per_band, std_per_band), f)
        else:
            with open(z_score_fp, 'rb') as f:
                mean_per_band, std_per_band = pickle.load(f)
        for data in [train_data, valid_data, test_data]:
            apply_z_norm(data, mean_per_band, std_per_band)
    return train_data, valid_data, test_data
````

````python
def create_tf_dataset(data, batch_size):
    # Assuming `data` is a list of tuples (feats_audio, feats_other, targets)
    dataset = tf.data.Dataset.from_tensor_slices(data)
    dataset = dataset.shuffle(buffer_size=1000).batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
    return dataset
````

````python
def create_tf_dataset(charts, batch_size):
    audio_features = []
    other_features = []
    targets = []
    
    for chart in charts:
        context_radius = 7
        nframes = len(chart.song_features)
        print(chart)
        
        for i in range(context_radius, nframes - context_radius):
            context = chart.song_features[i-context_radius:i+context_radius+1]
            audio_features.append(context)
            other_features.append(np.zeros(1))  # Adjust size based on your other features
            targets.append(1.0 if i in chart.onsets else 0.0)
    
    dataset = tf.data.Dataset.from_tensor_slices((
        tf.convert_to_tensor(audio_features, dtype=tf.float32),
        tf.convert_to_tensor(other_features, dtype=tf.float32),
        tf.convert_to_tensor(targets, dtype=tf.float32)
    ))
    
    return dataset.shuffle(1000).batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
````
